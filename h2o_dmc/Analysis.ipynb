```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import pyvibdmc as dmc
from pyvibdmc.analysis import *
from pyvibdmc.simulation_utilities import *
from pyvibdmc import potential_manager as pm
```

```python
import matplotlib
matplotlib.rcParams.update({'errorbar.capsize': 8})
```

```python
ps_sim = SimInfo('h2o_ps_dmc_output/h2o_nn_0_sim_info.hdf5')
vref_ps = ps_sim.get_vref()

nn_sim = SimInfo('h2o_nn_dmc_output/h2o_nn_0_sim_info.hdf5')
vref_nn = nn_sim.get_vref()
```

```python
fig = plt.figure(figsize=(12,6))

fig_ps = fig.add_subplot(121)
fig_ps.hlines(y=np.mean(vref_ps[500:,1]), xmin=0, xmax=1e4, color='crimson', zorder=10)
ytick_ps = np.arange(0,6000,1000)
ytick_ps = np.append(ytick_ps, np.mean(vref_ps[500:,1]))
fig_ps.set_yticks(ytick_ps)
fig_ps.plot(vref_ps[:,0],vref_ps[:,1])
fig_ps.set_xlabel('Time step (atomic units)')
fig_ps.set_ylabel('$V_{ref}$ (cm$^{-1}$)')
fig_ps.title.set_text('Partridge Schwenke')

fig_nn = fig.add_subplot(122)
fig_nn.hlines(y=np.mean(vref_nn[500:,1]), xmin=0, xmax=1e4, color='crimson', zorder=10)
ytick_nn = np.arange(0,6000,1000)
ytick_nn = np.append(ytick_nn, np.mean(vref_nn[500:,1]))
fig_nn.set_yticks(ytick_nn)
fig_nn.plot(vref_nn[:,0],vref_nn[:,1])
fig_nn.set_xlabel('Time step (atomic units)')
fig_nn.set_ylabel('$V_{ref}$ (cm$^{-1}$)')
fig_nn.title.set_text('Neural Network')

plt.tight_layout()
plt.show()
```

```python
v_mean_ps = []
v_mean_nn = []

for i in range(5):
    ps_sim = SimInfo(f'h2o_ps_dmc_output/h2o_nn_{i}_sim_info.hdf5')
    vref_ps = ps_sim.get_vref()
    v_mean_ps.append(np.mean(vref_ps[500:,1]))

    nn_sim = SimInfo(f'h2o_nn_dmc_output/h2o_nn_{i}_sim_info.hdf5')
    vref_nn = nn_sim.get_vref()
    v_mean_nn.append(np.mean(vref_nn[500:,1]))
```

```python
x = np.arange(5)+1
v_mean_ps = np.array(v_mean_ps)
v_mean_nn = np.array(v_mean_nn)

fig = plt.figure(figsize=(12,6))

fig1 = fig.add_subplot(121)
fig1.plot(x, v_mean_ps, marker='o', label='PS')
fig1.plot(x, v_mean_nn, marker='v', label='NN')
fig1.set_xlabel('N$^{th}$ Simulation')
fig1.set_ylabel('ZPE (cm$^{-1}$)')
fig1.legend()

fig2 = fig.add_subplot(122)
fig2.bar(['PS', 'NN'], [np.mean(v_mean_ps), np.mean(v_mean_nn)], yerr=[np.std(v_mean_ps), np.std(v_mean_nn)])
fig2.set_ylim((4620,4650))
fig2.set_ylabel('Mean ZPE (cm$^{-1}$)')

plt.tight_layout()
plt.show()
```

```python
cds_ps = []
dws_ps = []
cds_nn = []
dws_nn = []

for i in range(5):
    ps_sim = SimInfo(f'h2o_ps_dmc_output/h2o_nn_{i}_sim_info.hdf5')
    cds, dws = ps_sim.get_wfns(np.arange(500,10500,1000))
    cds_ps += list(cds)
    dws_ps += list(dws)

    nn_sim = SimInfo(f'h2o_nn_dmc_output/h2o_nn_{i}_sim_info.hdf5')
    cds, dws = nn_sim.get_wfns(np.arange(500,10500,1000))
    cds_nn += list(cds)
    dws_nn += list(dws)
```

```python
analyzer_ps = AnalyzeWfn(np.array(cds_ps))
analyzer_nn = AnalyzeWfn(np.array(cds_nn))
```

```python
hoh_angle_ps = analyzer_ps.bond_angle(atm1=0,
                                atm_vert=2,
                                atm3=1)
hoh_angle_nn = analyzer_nn.bond_angle(atm1=0,
                                atm_vert=2,
                                atm3=1)
```

```python
hoh_angle_ps = np.rad2deg(hoh_angle_ps)
hoh_angle_nn = np.rad2deg(hoh_angle_nn)
```

```python
hoh_histo_ps = analyzer_ps.projection_1d(attr=hoh_angle_ps,  # make a 1d histogram , x/y data
                                   desc_weights=np.array(dws_ps),
                                   bin_num=80,
                                   range=(60, 150))
hoh_histo_nn = analyzer_nn.projection_1d(attr=hoh_angle_nn,  # make a 1d histogram , x/y data
                                   desc_weights=np.array(dws_nn),
                                   bin_num=80,
                                   range=(60, 150))
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(hoh_histo_ps[:,0], hoh_histo_ps[:,1], label='PS')
ax.plot(hoh_histo_nn[:,0], hoh_histo_nn[:,1], label='NN')
ax.set_xlabel('HOH Angle (Degrees)')
ax.set_ylabel('$|\Psi|^2$')
ax.legend()

plt.show()
```
